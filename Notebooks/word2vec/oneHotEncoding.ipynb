```python
# Paso 1: Crear el vocabulario
vocabulario = ["gato", "perro", "pájaro", "pez"]

# Paso 2: Crear un diccionario que asigne un índice a cada palabra
indice_palabras = {palabra: i for i, palabra in enumerate(vocabulario)}

# Paso 3: Función para codificar una palabra
def one_hot_encode(palabra, vocabulario):
    vector = [0] * len(vocabulario)

    # Esta funcion marca con 1 la posición correspondiente a la palabra
    if palabra in indice_palabras:
        indice = indice_palabras[palabra]
        vector[indice] = 1
    return vector

# Ejemplo de un uso
print(one_hot_encode("gato", vocabulario))


# Paso 4: Imprimir todos los vectores one-hot
def imprimir_todos_one_hot(vocabulario):
    for palabra in vocabulario:
        vector = one_hot_encode(palabra, vocabulario)
        print(f"{palabra} → {vector}")

# Llamar a la función
imprimir_todos_one_hot(vocabulario)
```
